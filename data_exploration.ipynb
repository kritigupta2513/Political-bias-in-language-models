```python
import os
import pandas as pd
import json

# Define the path to the folder containing JSON files
folder_path = '/Users/kritigupta/mlp project/Article-Bias-Prediction/data/jsons'

# List all JSON files in the folder
json_files = [pos_json for pos_json in os.listdir(folder_path) if pos_json.endswith('.json')]

# Initialize an empty list to store data
data_list = []

# Loop through each JSON file and read its content
for json_file in json_files:
    file_path = os.path.join(folder_path, json_file)
    with open(file_path, 'r') as file:
        data = json.load(file)
        data_list.append(data)

# Convert the list of JSON data to a DataFrame
df = pd.DataFrame(data_list)

# # Display the DataFrame
# print(df)
```

```python
import torch
from transformers import RobertaModel, RobertaTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer, DataCollatorWithPadding
from peft import LoraConfig, get_peft_model
```

```python
peft_model_name = 'roberta-base-peft'
modified_base = 'roberta-base-modified'
base_model = 'roberta-base'
```

```python
tokenizer = RobertaTokenizer.from_pretrained(base_model)
```

```python
def preprocess(examples):
    tokenized = tokenizer(examples['text'], truncation=True, padding=True)
    return tokenized
```

```python
tokenized_dataset = train_dataset.map(preprocess, batched=True, remove_columns=["text"])
```

```python
tokenized_valid_dataset = valid_dataset.map(preprocess, batched=True,  remove_columns=["text"])
```

```python
tokenized_test_dataset = test_dataset.map(preprocess, batched=True,  remove_columns=["text"])
```

```python
num_labels = df.bias_text.nunique()
class_names = df.label.unique()
print(f"number of labels: {num_labels}")
print(f"the labels: {class_names}")
```

```python
id2label = {i: label for i, label in enumerate(class_names)}
```

```python
data_collator = DataCollatorWithPadding(tokenizer=tokenizer, return_tensors="pt")
```

```python
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy='steps',
    learning_rate=5e-5,
    num_train_epochs=1,
    per_device_train_batch_size=16,
)
```

```python
def get_trainer(model):
      return  Trainer(
          model=model,
          args=training_args,
          train_dataset=tokenized_dataset,
          eval_dataset=tokenized_valid_dataset,
          data_collator=data_collator,
      )
```

```python
full_finetuning_trainer = get_trainer(
    AutoModelForSequenceClassification.from_pretrained(base_model, id2label=id2label),
)

full_finetuning_trainer.train()
```
